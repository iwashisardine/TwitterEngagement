実行するとマイドライブのdataディレクトリにあるimport_file(=pre_data.csv)からtextだけを取得してexport_file(=pre_text.csv)としてdataディレクトリに出力する

```python
# run to get text data
import csv

number = 0

import_file = 'pre_data_ga_much2_new.csv'
export_file = 'pre_text_ga_much2.csv'

with open('/content/drive/MyDrive/data/' + import_file) as f:
    data = csv.reader(f)
    for datum in data:
        ex_text = [datum[0]]

        # textラベルも取り除く
        if not number == 0:
            with open('/content/drive/MyDrive/data/' + export_file, 'a', newline='') as f:
                writer = csv.writer(f)
                writer.writerow(ex_text)
        
        print(number)
        number += 1
```
