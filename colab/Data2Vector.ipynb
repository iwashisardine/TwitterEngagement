```python
# run to vectorize
import csv
from gensim.models.doc2vec import Doc2Vec

model = Doc2Vec.load("/content/drive/MyDrive/model/d2v_ga_1.model")
vec_size = 400

number = 1

import_file = 'pre_data_ga_new.csv'
export_file = 'vec_data_ga.csv'

with open('/content/drive/MyDrive/data/' + import_file) as f:
    data = csv.reader(f)
    for datum in data:
        if number == 1:
            text = []
            for i in range(vec_size):
                text.append(f'tx_feature{i+1}')
        else:
            nptext = model.docvecs[number-2]
            text = nptext.tolist()
        
        del datum[0]
        del datum[-1]
        #ful_datum = text
        ful_datum = text + datum

        with open('/content/drive/MyDrive/data/' + export_file, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(ful_datum)

        print(number)
        number += 1
```
