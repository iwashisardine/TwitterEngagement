```python
# run to load module
import csv
import datetime
import locale
import tqdm
```

```python
# run to modify time data and attach label
'''check here before running'''
import_path = '/content/drive/Shareddrives/NLP/data/vec_data_osusume_wiki.csv'   # csv that texts are vectored, id is removed
export_dir = '/content/drive/Shareddrives/NLP/data/'

export_path = export_dir + 'Rmod_osusume_wiki.csv'
number = 1

def day2num(day):
    if day == "Monday":
        return 1
    elif day == "Tuesday":
        return 2
    elif day == "Wednesday":
        return 3
    elif day == "Thursday":
        return 4
    elif day == "Friday":
        return 5
    elif day == "Saturday":
        return 6
    elif day == "Sunday":
        return 7

with open(import_path) as f:
    data = csv.reader(f)
    eng = 0
    for datum in tqdm.tqdm(data):
        if number == 1:
            datum.append("day")
            datum.append("hour")
            datum.append("minute")
            
            datum.insert(0, "engagement")
        else:
            time = datum[-6]
            dt = datetime.datetime.strptime(time, '%Y-%m-%d_%H:%M:%S')
            strday = dt.strftime('%A')
            day = day2num(strday)
            hour = dt.hour
            minute = dt.minute

            eng = int(datum[-2]) + int(datum[-3])

            datum.append(day)
            datum.append(hour)
            datum.append(minute)
            datum.insert(0, eng)

        del datum[-9]
        del datum[-6]
        del datum[-5]

        with open(export_path, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(datum)

        number += 1
```

```python
# run to convert into R_pattern1
# pattern1 = [engagement, [tx_feature(400)]]

import_path = '/content/drive/Shareddrives/NLP/data/data.Rmod.csv'
export_path = '/content/drive/Shareddrives/NLP/data/dataset.Rpat1.csv'

with open(import_path) as f:
    data = csv.reader(f)
    for datum in tqdm.tqdm(data):
        for i in range(6):
            del datum[-1]
        with open(export_path, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(datum)
```

```python
# run to convert into R_pattern2
# pattern2 = [engagement, [tx_feature(400)], follow, follower, day, hour, minute]

import_path = '/content/drive/Shareddrives/NLP/data/Rmod_ga.csv'
export_path = '/content/drive/Shareddrives/NLP/data/Rpat2_ga.csv'

with open(import_path) as f:
    data = csv.reader(f)
    for datum in tqdm.tqdm(data):
        del datum[-4]
        with open(export_path, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(datum)
```

```python
# run to convert into pattern3
# pattern3 = [engagement(label), [tx_feature(400)], ratio]

import_file = 'dataset.pat1_2.csv'
export_file = 'dataset.pat3.csv'

number = 1

with open('/content/drive/MyDrive/data/' + import_file) as f:
    data = csv.reader(f)
    for datum in data:
        if number == 1:
            datum.append("ratio")
        else:
            follow = int(datum[-5])
            follower = int(datum[-4])
            if follow == 0:
                ratio = 0
            else:
                ratio = follower / follow
            
            datum.append(ratio)

        for i in range(5):
            del datum[-2]

        with open('/content/drive/MyDrive/data/' + export_file, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(datum)

        print(number)
        number += 1
```
