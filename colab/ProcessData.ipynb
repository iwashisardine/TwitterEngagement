```python
# run to preprocess tweetdata
import csv
import neologdn
import re
import sys

number = 1

import_file = 'tweet_data_ga_much2.csv'
export_file = 'pre_data_ga_much2_new.csv'
with open('/content/drive/MyDrive/data/' + import_file) as f:
    data = csv.reader(f)
    for datum in data:
        ex_text = datum[0]
        
        oneline_text = ex_text.replace('\n', ' ')
        normalized_text = neologdn.normalize(oneline_text)

        text_without_url = re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-]+', '', normalized_text)

        non_bmp_map = dict.fromkeys(range(0x10000, sys.maxunicode + 1), ' ')
        text_without_emoji = text_without_url.translate(non_bmp_map)

        re_pattern = re.compile(pattern = "[✨⭕⭐⏩✊✅⬇️❗❤‍❄←↓↑→➡⚡❣✒①②③④⑤⑥⑦⑧⑨⑩❶❷❸❹❺❻❼❽❾➓⏰➰✌❓❌⚓⬛❕✋⚠✔⛰⌯⛵]")
        text_without_emoji = re_pattern.sub(r'', text_without_emoji)

        tmp = re.sub(r'(\d)([,.])(\d+)', r'\1\3', text_without_emoji)
        text_replaced_number = re.sub(r'\d+', '0', tmp)

        tmp = re.sub(r'[!-/:-@[-`{-~]', r' ', text_replaced_number)

        text_removed_symbol = re.sub(u'[■-♯]', ' ', tmp)
        
        print(f"data{number}:{text_removed_symbol}")

        datum[0] = text_removed_symbol

        with open('/content/drive/MyDrive/data/' + export_file, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(datum)

        number += 1
```

```python
# run to preprocess tweetdata
import csv
import neologdn
import re
import sys
import tqdm

number = 1

import_file = 'pre_data_ga_much2_new.csv'
export_file = 'pre_data_ga_mach2_tuika.csv'
with open('/content/drive/MyDrive/data/' + import_file) as f:
    data = csv.reader(f)

    data_list = [datum for datum in data]
    for i in tqdm.tqdm(range(len(data_list))):
        ex_text = data_list[i][0]

        re_pattern = re.compile(pattern = "[【】「」《》⋆⁑•﹏✖⁉『』““‥₍ ̨̡‾᷄꒫‾᷅ ̧̢₎·̩͙ᵕᴗ。、✍≧∇≦•̀… ﾟ0ﾟ´｀・ω´艸｀彡‼⛑✍❔…∠✖⚽✈▓█░▒✏⏬♻†]")
        text_without_brackets = re_pattern.sub(r'', ex_text)

        data_list[i][0] = text_without_brackets

        with open('/content/drive/MyDrive/data/' + export_file, 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(data_list[i])

        number += 1
```
